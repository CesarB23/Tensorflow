```python
from tensorflow.keras.models import load_model
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras

loaded_model = keras.models.load_model("modelo_completo.h5")
```

```python
img_size = 128

new_image = cv2.resize(cv2.imread(r'C:\Users\52221\Desktop\Programacion\ASE_2\imagenes de prueba\imagen prueba 1 corte axial.jpg'), (img_size, img_size))
print(new_image.shape)
new_image = np.expand_dims(new_image, axis=0)
new_image = new_image.astype('float32') / 255.0
prediction = loaded_model.predict(new_image)
predicted_class = np.argmax(prediction)

classes = {"0":"Glioma",
           "1":"Meningioma",
           "2":"No Tumor",
           "3":"Pituitario"}

# # Imprime la clase predicha
print("Clase predicha:", classes[f"{predicted_class}"])
```
